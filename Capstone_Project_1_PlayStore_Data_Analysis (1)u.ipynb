## 2. Handling Missing/NaN values:

```python
def playstore_info():
    info = pd.DataFrame(index = playstore_df.columns)
    info['Datatypes'] = playstore_df.dtypes
    info['Not null values'] = playstore_df.count()
    info['Null values'] = playstore_df.isnull().sum()
    info['% of null values'] = (info['Null values'] / (info['Not null values'] + info['Null values']))*100
    info['Unique count'] = playstore_df.nunique()
    return info
playstore_info()
```

**4). Rating: This column contains 1459 NaN values.**

```python
playstore_df[playstore_df['Rating'].isnull()]
```

The Rating column contains 1470 NaN values which accounts to apprximately 13.5% of the rows in the entire dataset.      

**Points to notice:**
1. It is not practical to drop these rows because by doing so, we will loose a large amount of data, which may impact the final quality of the analysis.                                                                                                       
2. The NaN values in this case can be imputed by the aggregate (mean or median) of the remaining values in the Rating column.

```python
# Finding mean and median in the Rating column excluding the NaN values.
mean_rating = playstore_df[~playstore_df['Rating'].isnull()]['Rating'].mean()
median_rating = playstore_df[~playstore_df['Rating'].isnull()]['Rating'].median()
print(f"The mean of the values of the 'Rating' column is {round(mean_rating,4)}")
print(f"The median of the values of the 'Rating' column is {median_rating}")
```

**Points to notice:**                                                                                                     
1. The mean of the average ratings (excluding the NaN values) comes to be 4.2.
2. The median of the entries (excluding the NaN values) in the 'Rating' column comes to be 4.3. From this we can say that 50% of the apps have an average rating of above 4.3, and the rest below 4.3.

Therefore, the median is a better representation of the majority of the values in the variable.
Hence we will impute the NaN values in the Rating column with its median.

```python
# Replacing the NaN values in the 'Rating' column with its median value 
playstore_df['Rating'].fillna(median_rating, inplace = True)
```

## 2. size:

```python
def convert_kb_to_mb(val):
  '''
  This function converts all the valid entries in KB to MB and returns the result in float datatype.
  '''
  try:
    if 'M' in val:
      return float(val[:-1])
    elif 'k' in val:
      return round(float(val[:-1])/1024, 4)
    else:
      return val
  except:
    return val
```

```python
playstore_df['Size'] = playstore_df['Size'].apply(lambda x: convert_kb_to_mb(x))
playstore_df.head()
```

```python
playstore_df['Size'] = playstore_df['Size'].apply(lambda x : str(x).replace('Varies with device', 'NaN') if 'Varies' in str(x) else x)
```

## Problem 1: Which are the top 5 most popular categories ?

### Chart 1: Bar chart

```python
def category_wise_apps(Category_name, App_name):

    popular_category = playstore_df.groupby(Category_name, as_index = False)[App_name].count().sort_values(App_name, ascending=False)

    sns.set(rc = {'figure.figsize':(15,10)})
    ax = sns.barplot(x = App_name, y = Category_name, data = popular_category)
    ax.set_xlabel('Number of Apps', fontsize = 15)
    for bars in ax.containers:
        ax.bar_label(bars)
```

```python
category_wise_apps('Category', 'App')
```

##### What is/are the insight(s) found from the chart?

From the above chart, we can conclude that the most popular categories are:
Family, Games, Tools, Business, Medical   

And the least popular categories are:

## Problem 2 : Which categories comprise the majoriy of installs?

### Chart 3: Bar chart

```python
def category_wise_installs(Category_name, column_name):

    popular_categories = playstore_df.groupby(Category_name, as_index = False)[column_name].sum().sort_values(by = 'Installs', ascending = False)
    popular_categories[column_name] = popular_categories[column_name].apply(lambda x : round(x/1000000000,2))
    
    sns.set(rc = {'figure.figsize':(15,10)})
    ax = sns.barplot(x = column_name, y = Category_name, data = popular_categories)
    ax.set_ylabel('Number of installs(in Billions)', fontsize = (15))
    ax.set_xlabel('Categories', fontsize = (15))
    ax.set_title('Top 5 popular categories with most number of installs(in Billions)', fontsize = (20))
    
    for bars in ax.containers:
        ax.bar_label(bars)
```

```python
category_wise_installs('Category', 'Installs')
```

##### What is/are the insight(s) found from the chart?

Based on the above chart, it is given that the categories which consists of most number of installs are:
**Games**, **Communication** and **Tools**.

## Problem 6: What is the most successfull size range based on the app count?

### Chart 7: Bar chart

```python
def size_range(size):
    if size < 1:
        return 'Below 1'
    elif size > 1 and size < 30:
        return '1-30'
    elif size >= 30 and size < 60:
        return '30-60'
    elif size >= 60 and size < 90:
        return '60-90'
    elif size >= 90 and size < 120:
        return '90-120'
```

```python
playstore_df['Size_Range'] = playstore_df['Size'].apply(size_range)
```

```python
playstore_df['Size_Range'].value_counts()
```

```python
playstore_df.head()
```

```python
def get_size_wise_count(column1, column2):   
    size_range_count = playstore_df.groupby(column1, as_index = False)[column2].count().sort_values(column2, ascending = False)
    sns.set(rc = {'figure.figsize' : (10,5)})

    ax = sns.barplot(x=column1, y=column2, data=size_range_count)
    for bars in ax.containers:
        ax.bar_label(bars)
    ax.set_xlabel('Size Range in MBs') 
    ax.set_ylabel('Number of Apps');
```

```python
get_size_wise_count('Size_Range', 'App')
```

#### What is/are the insight(s) found from the chart?

Based on the above chart, majority of the apps present in the playstore have size less than 30 MBs.

```python
def get_version_cat(val):
    if val.find('1') == 0:
        return 'version 1'
    elif val.find('2') == 0:
        return 'version 2'
    elif val.find('3') == 0:
        return 'version 3'
    elif val.find('4') == 0:
        return 'version 4'
    elif 'Varies' in val:
        return val
    else:
        return '6 or above'
```
